## Matplotlib example workflow

```python
# 1. import matplotlib and get it ready for plotting.
%matplotlib inline
import matplotlib.pyplot as plt

# 2 . Prepare some data
x = [1,2,3,4]
y = [11,22,33,44]

# 3. Setup plot
fig , ax = plt.subplots(figsize = (10,10))
# figsize refers the width and height of the figure.

# 4. Plot the dat
ax.plot(x,y)

# 5. Customize plot
ax.set(title = "Simple Plot",
       xlabel = "x-axis",
       ylabel = "y-axis")
# 6. Save and show
fig.savefig("Images/sample-plot.png")

# 7 . Show the plot
plt.show()
```

## Subplots
## Two options for subplots

Option 1

```python
# Subplots option 1  we use this.
fig , ((ax1,ax2), (ax3,ax4)) = plt.subplots(nrows=2,
                                            ncols=2,
                                            figsize=(10,5))
# Plot to each different axis
ax1.plot(x , x/2)
ax2.scatter(np.random.random(10) , np.random.random(10))
ax3.bar(nut_butter_prices.keys() ,nut_butter_prices.values())
ax4.hist(np.random.randn(1000))

# Display the graphs
plt.show()
```

```python
# Subplots option 2
fig, ax = plt.subplots(nrows = 2 ,
                       ncols = 2 ,
                       figsize = (10,5))
# Plot to each different indeex
ax[0,0].plot(x , x/2)
ax[0,1].scatter(np.random.random(10) , np.random.random(10))
ax[1,0].bar(nut_butter_prices.keys() ,nut_butter_prices.values())
ax[1,1].hist(np.random.randn(1000))

# Display the graphs
plt.show()
```

```python
import pandas as pd

car_sales = pd.read_csv("car-sales.csv")
car_sales
```

```python
# Step 1: Convert to string (so we can use .str.replace)
car_sales["Price"] = car_sales["Price"].astype(str)

# Step 2: Remove $, commas, and periods using regex
car_sales["Price"] = car_sales["Price"].str.replace(r'[\$,\.]', '', regex=True)

# Step 3: Convert to numeric (optional: use errors='coerce' to handle bad data)
car_sales["Price"] = pd.to_numeric(car_sales["Price"], errors='coerce')

# (Optional) Drop rows with missing prices
car_sales = car_sales.dropna(subset=["Price"])

# (Optional) Convert to int if no cents are needed
car_sales["Price"] = car_sales["Price"].astype(int)

car_sales
```

```python
type(car_sales["Price"][0])
```

```python
# Remove last two digits from each price by integer division
car_sales["Price"] = car_sales["Price"] // 100
car_sales
```

```python
car_sales["Sale Date"] = pd.date_range("1/1/2020" , periods=len(car_sales))
car_sales
```

```python
car_sales["Total Sales"] = car_sales["Price"].cumsum()
car_sales
```

```python
# Lets plot the total sale
car_sales.plot(x = "Sale Date" ,y = "Total Sales")
plt.show()
```

```python
# Plot scatterplot with odo and price col
car_sales.plot(x = "Odometer (KM)" ,y = "Price", kind = "scatter")   # here kind = scatter returns scatter plot
plt.show()
```

```python
# Plot bar graph with make and price col
car_sales.plot(x = "Make" ,y = "Price", kind = "bar")   # here kind = bar returns bar graph
plt.show()
```

### Which one should you use? (pyplot vs matplotlib OO method)

* When plotting something quickly , okay to use pyplot method.
* When plotting something more advanced, use th OO method

```python
heart_disease.head()
```

```python
over_50 = heart_disease[heart_disease["age"] > 50 ]
over_50
```

```python
over_50.head()
```

```python
len(over_50)
```

```python
# Pyplot method aka plotting directly from a pandas dataframe.
over_50.plot(kind="scatter",
             x="age",
             y="chol",
             c="target")     # here the c referes color so color it on the basis of target column.
plt.show()
```

```python
# OO method mixed with pyplot method
fig , ax = plt.subplots()
over_50.plot(kind = 'scatter',
             x = 'age',
             y = 'chol',
             c = 'target',
             ax = ax)    # this ax referes the axis we created that is from fig,ax
plt.show()
```

```python
fig , ax = plt.subplots()
over_50.plot(kind = 'scatter',
             x = 'age',
             y = 'chol',
             c = 'target',
             ax = ax) 
ax.set_xlim([45 , 100])
plt.show()
```

```python
## OO method from scratch
fig , ax = plt.subplots(figsize=(10,6))

# Plot the data
scatter = ax.scatter(x = over_50["age"],
                     y = over_50["chol"],
                     c = over_50["target"])

# Customize the plot
ax.set(title = "Heart Disease and Cholesterol Level",
       xlabel = "Age",
       ylabel = "Cholesterol")

# Add a legend
ax.legend(*scatter.legend_elements() , title="Target")

# Add a horizontal line at the mean cholesterol level
ax.axhline(y=over_50["chol"].mean(numeric_only=True),
           linestyle="--", 
           color="red")


plt.show()
```

```python
## OO method from scratch 2 
fig , ax = plt.subplots(figsize = (10,6))

# Plot the data
scatter = ax.scatter(x = over_50['age'],
                     y = over_50['thalach'],
                     c = over_50['target'])

# Customize the plot
ax.set(title = 'Max Heart Rate and Age',
       xlabel = 'Age',
       ylabel = 'Max Heart Rate'
      )

# Add a legend
ax.legend(*scatter.legend_elements(), title = 'Target')

plt.show()
```

```python
## OO method for subplot of chol , age , and thalach

fig ,(ax0 , ax1) = plt.subplots(nrows = 2 ,
                                ncols = 1,
                                figsize=(10,10),
                                sharex = True)    # Now both axx0 and ax1 share xlabel = Age

# Plot the data to ax0
scatter = ax0.scatter(x = over_50["age"],
                      y = over_50["chol"],
                      c = over_50["target"])

# Customize the plot of ax0
ax0.set(title = "Heart Disease and Cholesterol Level",
        ylabel = "Cholesterol")

# Add a legend to ax0
ax0.legend(*scatter.legend_elements() , title="Target")

# Add a horizontal line at the mean cholesterol level to ax0
ax0.axhline(y=over_50["chol"].mean(numeric_only=True),
            linestyle="--", 
            color="red")



# Plot the data to ax1
scatter = ax1.scatter(x = over_50["age"],
                      y = over_50["thalach"],
                      c = over_50["target"])

# Customize the plot of ax1
ax1.set(title = "Heart Disease and Max Heart Rate",
        xlabel = "Age",
        ylabel = "Max Heart Rate")

# Add a legend to ax1
ax1.legend(*scatter.legend_elements() , title="Target")

# Add a horizontal line at the mean thalach (max heart rate) to ax1
ax1.axhline(y=over_50["thalach"].mean(numeric_only=True),
            linestyle="--", 
            color="blue")

# Add title to the figure
fig.suptitle("Heart Disease Analysis",
             fontsize = 16,
             fontweight = "bold"
            )

plt.show()
```

```python
%matplotlib inline

# Customizing the y and x axis limitations
## OO method for subplot of chol , age , and thalach

fig ,(ax0 , ax1) = plt.subplots(nrows = 2 ,
                                ncols = 1,
                                figsize=(10,10),
                                sharex = True)    # Now both axx0 and ax1 share xlabel = Age

# Plot the data to ax0
scatter = ax0.scatter(x = over_50["age"],
                      y = over_50["chol"],
                      c = over_50["target"],
                      cmap = 'winter')

# Customize the plot of ax0
ax0.set(title = "Heart Disease and Cholesterol Level",
        ylabel = "Cholesterol")

# Change the x axis limit
ax.set_xlim([50 , 80])

# Add a legend to ax0
ax0.legend(*scatter.legend_elements() , title="Target")

# Add a horizontal line at the mean cholesterol level to ax0
ax0.axhline(y=over_50["chol"].mean(numeric_only=True),
            linestyle="--", 
            color="red")



# Plot the data to ax1
scatter = ax1.scatter(x = over_50["age"],
                      y = over_50["thalach"],
                      c = over_50["target"],
                      cmap = 'winter')

# Customize the plot of ax1
ax1.set(title = "Heart Disease and Max Heart Rate",
        xlabel = "Age",
        ylabel = "Max Heart Rate")

# Change ax1 x axix limits and y axis limits
ax1.set_xlim([50 , 80])
ax1.set_ylim([60 , 200])

# Add a legend to ax1
ax1.legend(*scatter.legend_elements() , title="Target")

# Add a horizontal line at the mean thalach (max heart rate) to ax1
ax1.axhline(y=over_50["thalach"].mean(numeric_only=True),
            linestyle="--", 
            color="blue")

# Add title to the figure
fig.suptitle("Heart Disease Analysis",
             fontsize = 16,
             fontweight = "bold"
            )

plt.show()
```
